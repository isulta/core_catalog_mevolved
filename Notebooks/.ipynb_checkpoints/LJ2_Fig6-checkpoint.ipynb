## Perform spatial matching of unmatched halos in subhalo and core catalogs (to get perfect match in mass bin)

```python
sod_cols = ['fof_halo_tag', 'sod_halo_cdelta', 'fof_halo_mass', 'sod_halo_mass',
            'fof_halo_center_x','fof_halo_center_y','fof_halo_center_z',
            'sod_halo_mean_x', 'sod_halo_mean_y', 'sod_halo_mean_z',
            'sod_halo_radius']
```

```python
sod_SV0_sh = gio_read_dict('/home/isultan/data/LJDS/M200/m000p-499.haloproperties', sod_cols)
sod_SV0_cc = gio_read_dict('/home/isultan/data/LJDS/M200_cc/m000p-499.haloproperties', sod_cols)

sod_sh = sod_SV0_sh
sod_cc = sod_SV0_cc
```

```python
assert np.all(np.isin(sh['fof_halo_tag'], sod_sh['fof_halo_tag'])), 'FOF halo tags exist in sh that are not in sod_sh.'
assert np.all(np.isin(cc['fof_halo_tag'], sod_cc['fof_halo_tag'])), 'FOF halo tags exist in cc that are not in sod_cc.'
```

```python
fht_smatch_cc, fht_smatch_sh = sod_spatial_match(sod_cc, sod_sh, M1, M2, SHMLM_SV.BOXSIZE)
assert not np.any(np.isin(fht_smatch_cc, sh)), 'New matched sod_cc tag already existed in sh'
```

```python
sh['fof_halo_tag'] = replace_elems(sh['fof_halo_tag'], fht_smatch_sh, fht_smatch_cc)
```
