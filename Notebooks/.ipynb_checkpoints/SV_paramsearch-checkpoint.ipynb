```python
for R in [R_over1000p, R_12_13, R_13_14, R_14_15]:
    plt.pcolormesh(zeta_arr, A_arr, R, cmap=plt.cm.jet)
    cb = plt.colorbar()
    cb.set_label('2-norm relative error')
    plt.xlabel('$\zeta$')
    plt.ylabel('A')
    print cb.vmin, cb.vmax
    plt.close()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True, sharey=True, gridspec_kw={'hspace': 0, 'wspace': 0}, figsize=[4.8*2,4.8*1.5], dpi=150)
im = ax1.pcolormesh(zeta_arr, A_arr, R_12_13, cmap=plt.cm.jet, vmin=0.5182335656856432, vmax=28.15563072547869)
ax2.pcolormesh(zeta_arr, A_arr, R_13_14, cmap=plt.cm.jet, vmin=0.5182335656856432, vmax=28.15563072547869)
ax3.pcolormesh(zeta_arr, A_arr, R_14_15, cmap=plt.cm.jet, vmin=0.5182335656856432, vmax=28.15563072547869)
ax4.pcolormesh(zeta_arr, A_arr, R_over1000p, cmap=plt.cm.jet, vmin=0.5182335656856432, vmax=28.15563072547869)

ax1.set_title( '{} $\leq$ log(M/$h^{{-1}}M_\odot$)$\leq$ {}'.format(12,13), pad=-16, x=0.5, color='w' )
ax2.set_title( '{} $\leq$ log(M/$h^{{-1}}M_\odot$)$\leq$ {}'.format(13,14), pad=-16, x=0.5, color='w' )
ax3.set_title( '{} $\leq$ log(M/$h^{{-1}}M_\odot$)$\leq$ {}'.format(14,15), pad=-16, x=0.5, color='w' )
ax4.set_title( '$M \ge1000$ parts.', pad=-16, x=0.5, color='w' )

fig.subplots_adjust(right=0.83)
cbar_ax = fig.add_axes([0.85, 0.15, 0.03, 0.7])
fig.colorbar(im, cax=cbar_ax)
plt.ylabel('2-norm relative error')

fig.text(0.5,0.08, r'$\zeta$', ha="center", va="center")
fig.text(0.08,0.5, r'A', ha="center", va="center", rotation=90)
```
