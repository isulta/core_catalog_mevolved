```python
def hostbins(M1_th, M2_th, verbose=False, Nhalos=10**4, minHalosFlag=False):
    massmask = (cc['infall_tree_node_mass'][centrals_mask] >= M1_th)&(cc['infall_tree_node_mass'][centrals_mask] < M2_th)
    asort = np.argsort(cc['infall_tree_node_mass'][centrals_mask][massmask])
    if verbose:
        print 'Total host halos within mass threshold', np.sum(massmask)

    cim = cc['infall_tree_node_mass'][centrals_mask][massmask][asort]
    cct = cc['core_tag'][centrals_mask][massmask][asort]   

    if minHalosFlag:
        Nhalos += np.sum(cim[Nhalos:]==cim[Nhalos-1])
    if verbose:
        print 'Number of host halos in bin', Nhalos
        print ''
    return cct[:Nhalos], cim[:Nhalos]
#     asort2 = np.argsort(cct[:Nhalos])
    
#     Mmatrix = np.full((101,Nhalos), -1, np.float32)
#     Mmatrix[0] = cim[:Nhalos][asort2]
#     return cct[:Nhalos][asort2], Mmatrix
```

```python
%%time
Mbins = [12,12.5]#[13.0, 13.5, 14.0, 14.5]
cctbins = []
cimbins = []
for lM in Mbins:
    cct, cim = hostbins(10**lM, 10**(lM+0.5), verbose=True, minHalosFlag=True)
    cctbins.append(cct)
    cimbins.append(cim)
```

```python
%%time
cctbins = np.concatenate(cctbins)
cimbins = np.concatenate(cimbins)
asortbins = np.argsort(cctbins)
cctbins_sorted = cctbins[asortbins]

Mmatrix = np.full((101,len(cctbins)), -1, np.float32)
Mmatrix[0] = cimbins[asortbins]
```

```python
%%time
for row, s in enumerate(SHMLM.steps[::-1]):#[499, 487]):
    print s
    if row == 0:
        continue
    cc = h5_read_dict('output_LJDS_localhost_dtfactor_0.5/m000p-{}.corepropertiesextend.hdf5'.format(s), 'coredata')
#     cc = h5_read_dict('/home/isultan/data/LastJourney/LJ_CoreCatalog_004_reduced/m000p-{}.coreproperties_complete.hdf5'.format(s), 'coredata')
    centrals_mask = cc['central'] == 1
    
    cc_ct_cm = cc['core_tag'][centrals_mask]
    cct_idx = np.isin( cctbins_sorted, cc_ct_cm )
    isin = np.isin( cc_ct_cm, cctbins_sorted )
    asort = np.argsort(cc_ct_cm[isin])
    assert np.sum(isin) == np.sum(cct_idx), 'Duplicate core tags found.'
    Mmatrix[row,cct_idx] = cc['infall_tree_node_mass'][centrals_mask][isin][asort]
```
