```python
def plt_latex():
	# Latex for plt
	from matplotlib import rc
	rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
	rc('text', usetex=True)
plt_latex()
```

```python
def delta_vir_G10(z):
    alpha = 0.7076
    beta = 0.4403
    return 9 * np.pi**2 * ( 1+Omega(z)**beta - alpha*(1-Omega(z)) )
```

```python
xarr = np.linspace(0,10,500)
plt.figure(dpi=100)
plt.plot(xarr, delta_vir_G10(xarr), label='Giocoli et al. 2010')
plt.plot(xarr, delta_vir(xarr), label='Bryan and Norman 1998')
plt.legend();
plt.title(r'$\Delta_{vir}(z)$ fitting functions')
plt.xlabel(r'$z$')
plt.ylabel(r'$\Delta_{vir}$');
```

```python
xarr = np.linspace(0,10,500)
plt.figure(dpi=100)
plt.plot(xarr, np.abs((delta_vir_G10(xarr)-delta_vir(xarr)))/( (delta_vir_G10(xarr)+delta_vir(xarr))/2 )*100)
# plt.legend();
plt.title(r'$\Delta_{vir}(z)$ fitting functions')
plt.xlabel(r'$z$')
plt.ylabel(r'\% diff. $\Delta_{vir}$ ');
```
